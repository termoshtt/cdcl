```python
import pandas
import matplotlib.pyplot as plt
import json
from glob import glob
from pathlib import Path
```

```python
for report in glob("./report_*.json"):
    with open(report) as f:
        data = json.load(f)
    label = Path(report).stem.removeprefix("report_")
    plt.plot(data["sorted_elapsed_times"], label=label)
plt.xlabel("# of instances")
plt.ylabel("Elaplsed time (ms)")
plt.legend()
```
